```python
# dummy data generator for testing
import random
import time
import pandas as pd
df = pd.read_csv('lifestyle_sustainability_data.csv')
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
categorical_df = df.select_dtypes(include=['object'])
```

```python
numeric_df.columns
```

```python
categorical_df.columns
```

```python
import plotly.express as px
# Veriyi HomeSize'a göre sıralama
df = df.sort_values(by='HomeSize')

# Çizgi grafiği oluşturma
fig = px.line(df, x='HomeSize', y='MonthlyElectricityConsumption', title='Energy Consumption vs Home Size')
fig.show()
```

```python

fig = px.parallel_categories(categorical_df)
fig.show()
```

```python
fig = px.bar(df, x='DisposalMethods', y='Age', labels={'DisposalMethods':'Disposal Methods', 'Age':'Age'}, title='Disposal Methods vs Age', color='DisposalMethods')
fig.show()
```

```python
#pie chart
fig = px.pie(df, names='HomeType', title='Types of Home')
fig.show()
#dispodsal methods vs home type
fig = px.bar(df, x='HomeType', y='DisposalMethods', labels={'DisposalMethods':'Disposal Methods', 'HomeType':'Home Type'}, title='Disposal Methods vs Home Type', color='HomeType')
fig.show()
```

```python
fig = px.histogram(df, x='HomeSize', title='Home Size Distribution', labels={'HomeSize':'Home Size'}, color='HomeSize', marginal='box', hover_data=df.columns, nbins=40, barmode='overlay')
fig.show()
```

```python
fig = px.box(df, x='HomeType', y='HomeSize', title='Home Size Distribution by Home Type', labels={'HomeSize':'Home Size', 'HomeType':'Home Type'}, color='HomeType')
fig.show()
```

```python
#density plot
fig = px.density_contour(df, x='HomeType', title='Home Size')
fig.show()
```

```python
fig = px.violin(df, x='HomeType', y='Age', title='Home Type Distrubition by age', labels={'Age':'Age', 'HomeType':'Home Type'}, color='HomeType')
fig.show()
```

```python
fig = px.scatter(df, x='HomeSize', y='Age', title='Home Size vs Age', labels={'HomeSize':'Home Size', 'Age':'Age'}, color='HomeType')
fig.show()
```

```python
fig = px.scatter_3d(df, x='MonthlyElectricityConsumption', y='Age', z='HomeSize', title='Monthly Electricity Consumption vs Age vs Environmental Awareness', labels={'MonthlyElectricityConsumption':'Monthly Electricity Consumption', 'Age':'Age', 'EnvironmentalAwareness':'Environmental Awareness'}, color='Age')
fig.show()
```

```python
import plotly.figure_factory as ff

colors = [ '#7FA6EE', '#B8F7D4']

fig = ff.create_distplot(
    [df['HomeSize']], 
    ['Home Size'], 
    bin_size=100, 
    show_rug=False, 
    show_curve=True, 
    show_hist=True, 
    colors=colors
)
fig.show()
```

```python
# heat map of df correlation with ployly

fig = px.imshow(numeric_df.corr(), title='Correlation Heatmap', labels=dict(color='Correlation'), x=numeric_df.columns, y=numeric_df.columns, color_continuous_scale='Viridis')
fig.show()
```
