```python
import pandas as pd

df= pd.read_csv('data.csv')

df.head()
```

```python
df = df.select_dtypes(include=['int64','float64'])
```

```python
df
```

```python
def can_be_ranked(df):
    rankable = {}
    for column in df.columns:
        if pd.api.types.is_numeric_dtype(df[column]):
            rankable[column] = True
        else:
            rankable[column] = False
    return rankable
rankable_columns = can_be_ranked(df)
rankable_columns

```

```python

# Example of the Mann-Whitney U Test
from scipy.stats import mannwhitneyu
df.dropna(inplace=True)
numeric_df = df.select_dtypes(include=['number'])

for col1 in numeric_df.columns:
    for col2 in numeric_df.columns:
        if col1 != col2:  
            data1 = numeric_df[col1]
            data2 = numeric_df[col2]
            stat, p = mannwhitneyu(data1, data2)
            print(f'Statistics for {col1} vs {col2}: stat=%.3f, p=%.3f' % (stat, p))
            if p > 0.05:
                print('Probably the same distribution')
            else:
                print('Probably different distributions')
```

```python
# check if df_numeric can be paired
def can_be_paired(df):
    pairable = {}
    for column in df.columns:
        if pd.api.types.is_numeric_dtype(df[column]):
            pairable[column] = True
        else:
            pairable[column] = False
    return pairable
pairable_columns = can_be_paired(numeric_df)
```

```python
pairable_columns
```

```python

# Example of the Wilcoxon Signed-Rank Test
from scipy.stats import wilcoxon

for col1 in pairable_columns:
    for col2 in pairable_columns:
        if col1 != col2:
            if pairable_columns[col1] and pairable_columns[col2]:
                data1 = numeric_df[col1]
                data2 = numeric_df[col2]
                stat, p = wilcoxon(data1, data2)
                print(f'Statistics for {col1} vs {col2}: stat=%.3f, p=%.3f' % (stat, p))
                if p > 0.05:
                    print('Probably the same distribution')
                else:
                    print('Probably different distributions')
```

```python
from scipy.stats import kruskal

for col1 in numeric_df.columns:
    for col2 in numeric_df.columns:
        if col1 != col2:
            data1 = numeric_df[col1]
            data2 = numeric_df[col2]
            stat, p = kruskal(data1, data2)
            print(f'Statistics for {col1} vs {col2}: stat=%.3f, p=%.3f' % (stat, p))
            if p > 0.05:
                print('Probably the same distribution')
            else:
                print('Probably different distributions')
```

```python
from scipy.stats import friedmanchisquare

for col1 in pairable_columns:
    for col2 in pairable_columns:
        for col3 in pairable_columns:
            if col1 != col2 and col1 != col3 and col2 != col3:
                if pairable_columns[col1] and pairable_columns[col2] and pairable_columns[col3]:
                    data1 = numeric_df[col1]
                    data2 = numeric_df[col2]
                    data3 = numeric_df[col3]
                    stat, p = friedmanchisquare(data1, data2, data3)
                    print(f'Statistics for {col1} vs {col2} vs {col3}: stat=%.3f, p=%.3f' % (stat, p))
                    if p > 0.05:
                        print('Probably the same distribution')
                    else:
                        print('Probably different distributions')
```
